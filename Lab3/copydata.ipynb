```python
import csv
f=open('cigarettes.csv','rb')
reader=csv.reader(f)
nicotine=[]
tar=[]
for row in reader:
    a=row[0]
    b=a.replace("\t", " ")
    c=b.split()
    d=c[1:]
    e=map(float,d)
    
    nicotine.append(e[1])
    tar.append(e[0])
    
print nicotine 
print tar
f.close()
```

```python
from pylab import*
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Tar and Nicotine in different brands')
ax.set_xlabel('Nicotine')
ax.set_ylabel('Tar')
scatter(tar,nicotine)
savefig('cig_scatter.png')
```
